**II. Marco teórico.**

**Transformada de Laplace.**

La transformada de Laplace emerge como una herramienta matemática fundamental en el análisis de señales y sistemas continuos. Proporciona una perspectiva poderosa al permitir la representación de sistemas lineales en el dominio del tiempo como funciones de transferencia en el dominio de Laplace, que opera en la frecuencia compleja $s$. Este enfoque facilita un análisis detallado y profundo de la respuesta y el comportamiento de estos sistemas. [1]

La transformada de Laplace utiliza como base la señal $e^{st}$, donde $s = σ+jw_{0}$ y puede interpretarse como una generalización de la transformada de Fourier.


Mediante la aplicación de la transformada de Laplace, es posible obtener la función de transferencia de un sistema lineal invariante en el tiempo y causal en el dominio del tiempo continuo. Esta función de transferencia proporciona una descripción integral de las propiedades del sistema, lo que permite un análisis detallado de su comportamiento. Además, a partir de la función de transferencia, es posible calcular algebraicamente la respuesta de salida del sistema para una amplia gama de señales de entrada, lo que facilita la predicción y comprensión del comportamiento del sistema en diferentes escenarios.

Por ejemplo, consideremos un sistema con una función de transferencia $H(s)$ y una señal de entrada escalón unitario con una transformada de Laplace $R(s) = \frac{1}{s}$. En este caso, la transformada de Laplace de la respuesta del sistema es $Y(s) = \frac{H(s)}{s}$. Para obtener la señal de salida en el dominio del tiempo, podemos aplicar la transformada inversa de Laplace a $Y(s)$. [2]

La transformada de Laplace de una señal $x(t)$ es:

$$X(S)=\int_{-\infty}^{\infty}x(t)e^{-st}dt$$

Los **polos** son los valores de $s$ que hacen que la transformada tienda a infinito, o en otras palabras, la raíces del denominador de la transformada de Laplace.

Por otro lado, los **ceros** son los valores de $s$ que hacen que la transformada tienda a cero, o en otras palabras, las raíces del numerador de la transformada de Laplace.

Las transformadas de Laplace tienen sus regiones de convergencia las cuales son comúnmente conocidas como ROC, los polos limitan a la ROC y no están contenidos en ella.

**Transformada inversa de Laplace.**

La determinación matemática de la ecuación de síntesis de la transformada de Laplace implica el cálculo de la integral de contorno en el plano complejo:

$$x(t)=\frac{1}{2\pi j}\int_{\sigma-jw}^{\sigma+jw}X(s)e^{st}ds$$

La resolución de la ecuación para obtener la transformada inversa de Laplace puede ser bastante complejo y poco común en la práctica. En su lugar, se recurre a procedimientos algebraicos que aprovechan los pares básicos de transformadas de Laplace y las propiedades inherentes de esta transformación. Estos métodos permiten calcular la transformada inversa de Laplace de una señal de manera más eficiente y efectiva, simplificando así el análisis de sistemas y señales en el dominio del tiempo continuo.

**Librería SymPy de Python.**

Sympy es una librería que puede utilizarse para calcular transformadas de Laplace en Python ya que proporciona la función *sym.laplace_transform(ft,t,s)* para expresiones $f(t)$ con términos simples. Esta función calcula la transformada de Laplace unilateral, donde t está en el intervalo $[0,∞)$, lo que implica entradas tipo causal con $t>0$, o términos de la forma $μ(t)$ desplazados hacia la derecha, $μ(t-1)$.

Al definir los símbolos $t$ y $s$, se establece la expresión correspondiente de $f(t)$ para determinar $F(s)$. [3]


**III. Procedimiento.**

**Transformada de Laplace.**

```python
#Definiendo una señal.
import sympy as sym
t = sym.Symbol('t')
s = sym.Symbol('s')
x = sym.cos(2*t)*sym.Heaviside(t)
p = sym.plotting.plot(x)
x
```

```python
#Cálculo de la transformada de Laplace.
X = sym.laplace_transform(x, t, s, noconds=True)
X
```

1. Determine la ubicación de polos y ceros y regiones de convergencia de las transformadas de Laplace de las señales que fueron definidas como objetos simbólicos.

a) Para $x_{1}(t) = sen(2 \pi t) u(t)$.

Al simplificar la transformada de Laplace de esta señal se obtiene:

$$X_{1}(s)= \frac{-2\pi i}{s^{2}+4\pi^{2}}$$

Los **polos** de $X_{1}(s)$ son las raíces del denominador, esto significa que los polos están $s = 2\pi i$ y $s = -2\pi i$.

Los **ceros**, dado que en el numerador no hay términos que puedan hacer que $X_{1}(s)$ se anule, no existen.

```python
# @title ROC de $X_{1}(S)$
import numpy as np
import matplotlib.pyplot as plt

real_values = np.linspace(-10, 10, 400) #rango de valores para la parte real de s
imag_values = np.linspace(-10, 10, 400) #rango de valores para la parte imaginaria de s
s_values = [r + 1j*i for r in real_values for i in imag_values]
ROC_real = [-10, 0] #región de convergencia para Re(s) < 0
ROC_imag = np.linspace(-10, 10, 1000)
fig, ax = plt.subplots(figsize=(8, 6)) #figura y eje
ax.fill_betweenx(ROC_imag, ROC_real[0], ROC_real[1], color='pink', label='Región de Convergencia') #gráfico de la ROC
ax.axhline(0, color='black',linewidth=0.5) #configuración de ejes
ax.axvline(0, color='black',linewidth=0.5)
ax.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
ax.set_xlabel('Re(s)')
ax.set_ylabel('Im(s)')
ax.set_title('Región de Convergencia (ROC)')
ax.set_xlim([-10, 2]) #límites de los ejes
ax.set_ylim([-10, 10]) #límites de los ejes
ax.legend() #mostrar la leyenda
plt.show()
```

La región de convergencia (ROC) de una transformada de Laplace es donde los polos están ubicados en el plano $s$. En este caso, los polos son $s = 2\pi i$ y $s = - 2\pi i$, con partes  reales iguales a cero y partes imaginarias diferentes de cero. La región de convergencia es el conjunto de valores
$s$ donde la parte real es menor que 0 ($Re(s)< 0$), excluyendo los ejes imaginarios. La gráfica muestra esta región como el área sombreada en rosado.

b) Para $x_{2}(t) = [u(𝑡 + 2) − 𝑢(𝑡 − 6)]$

Al simplificar la transformada de Laplace de esta señal se obtiene:

$$X_{2}(s)=\frac{1-e^{-6s}}{s}$$

Los **polos** de $X_{2}(s)$ son las raíces del denominador, esto significa que los polos están $s = 0$.

Los **ceros** se determinan igualando el numerador a cero y despejando la $s$, resolviendo se tiene:

$$e^{-6s}=1$$

$$-6s=\ln (1)$$

$$-6s=0$$

Finalmente se tiene que $s = 0$, por lo que en $s=0$ se encuentra el cero de la función.

```python
# @title ROC de $X_{2}(S)$
real_values = np.linspace(-10, 10, 400) #rango de valores para la parte real de s
imag_values = np.linspace(-10, 10, 400) #rango de valores para la parte imaginaria de s
s_values = [r + 1j*i for r in real_values for i in imag_values]
ROC_real = [0, 10] #región de convergencia para Re(s) > 0
ROC_imag = np.linspace(-10, 10, 1000)
fig, ax = plt.subplots(figsize=(8, 6)) #figura y eje
ax.fill_betweenx(ROC_imag, ROC_real[0], ROC_real[1], color='pink', label='Región de Convergencia') #gráfico de la ROC

ax.axhline(0, color='black',linewidth=0.5) #configuración de ejes
ax.axvline(0, color='black',linewidth=0.5)
ax.grid(color = 'gray', linestyle = '--', linewidth = 0.5)

ax.set_xlabel('Re(s)')
ax.set_ylabel('Im(s)')
ax.set_title('Región de Convergencia (ROC)')

ax.set_xlim([-2, 10]) #límites de los ejes
ax.set_ylim([-10, 10]) #límites de los ejes

ax.legend() #mostrar la leyenda

plt.show()
```

Un sistema LTI es estable si y solo si la ROC de su función de transferencia $H(s)$ incluye el eje $jw$. A partir de esta información se mostrarán las gráficas de la ROC de cada sistema y se determinará si son estables o no.

Para a) $H_{1}(s)= \frac{1}{s^{2}+4s}$
