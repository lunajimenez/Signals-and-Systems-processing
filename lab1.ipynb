**II. Marco teórico**

En esta sección se presentan conceptos importantes para el entendimiento de este laboratorio e inicio de curso.

Empecemos por lo más importante, ¿qué es una señal? ¿qué es un sistema?

Señal: una señal es una fluctuación de cualquier variable virtual o física. Algunos ejemplos pueden ser el voltaje, corriente, precio del dólar, la presión, la temperatura, etc. [1]

La base de los sistemas de comunicaciones electrónicas son componentes que procesan esencialmente señales digitales o analógicas. Una señal analógica fluctúa en el tiempo de forma continua.

Cuando se detectan señales analógicas en el entorno, un transductor las convierte en señales eléctricas para poder procesarlas electrónicamente.

Por otro lado, las señales digitales son discretas (valores finitos) en tiempo y amplitud, esto significa que la señal solo puede tomar dos valores, 0 o 1, en predeterminados intervalos de tiempo. [2]

Sistema: es aquel dispositivo o conjunto de dispositivos, ya bien pueden ser físicos o virtuales que procesan y/o transforman una(s) señal(es) en otra(s). Por ejemplo, los microprocesadores, transductores, entre otros. [1]

Los sistemas electrónicos se basan en circuitos interconectados para realizar tareas predeterminadas. Ejemplos habituales de este tipo de sistemas son los sistemas empotrados, los equipos de audio y vídeo, los dispositivos de medición y los sistemas de alarma, entre otros. [3]

**Punto #1: Ubicación Inicial de Satélites: variables complejas**

Ubicación central de la red de satélites y su gráfica correspondiente.

```python
#Para la posición central del satélite
import numpy as np #importamos la librería numpy
from matplotlib import pyplot as plt #Importamos de matplotlib un parámetro para graficar la posición del satélite
posicion_central = 15j-np.exp((1j*3*np.pi)) #Definimos la posición central de una red de satélites
print(posicion_central) #Mostramos la posición central
z=posicion_central #Llamamos z a la posición central
z.real #Devuelve la parte real de la variable z
z.imag #Devuelve la parte imaginaria de la variable z
plt.xlabel('Real'), plt.ylabel('Imaginario') #Definimos los ejes de la gráfica
plt.title('Posición central de una red de satélites') #Definimos el título de la gráfica
plt.plot(z.real, z.imag, 'o', color='purple')
plt.grid() #Ponemos una malla detrás de la gráfica
plt.show
```

**Punto #2. Trayectoria de satélites: Vectores y matrices**

Gráfico de la función que define la trayectoria del satélite.

```python
#Para la trayectoria de los satélites
#Debemos definir la función que describe la trayectoria del satélite y el rango de esa función
import numpy as np
from matplotlib import pyplot as plt
x = np.arange(-15,15+0.5,0.5) #arreglo que contiene el rango de interés
y = 3*x**2+7*x+2 #función trayectoria del satélite en función de x
plt.plot(x,y,color='purple') #graficaremos la trayectoria
plt.grid()
plt.xlabel('x'), plt.ylabel('y') #definimos los ejes del gráfico
plt.title('Trayectoria de los satélites') #definimos el título del gráfico
plt.show()
```

**Punto #3. Determinando Posición: Indexando vectores y matrices**



```python
#Determinando la posición de un satélite en un lapso de 24 horas
#Su posición puede ser en orbita 1 o en tierra (0)
import numpy as np
from matplotlib import pyplot as plt
t = np.arange(0,25) #arreglo que contiene el rango de interés
w = np.concatenate([np.zeros(5),np.ones(15),np.zeros(5)]) #arreglo para representar a la señal en función de t
plt.figure()
plt.plot(t,w,color='purple')
plt.grid()
plt.xlabel('t'), plt.ylabel('w') #definimos los ejes del gráfico
plt.title('Posición de un satélite') #definimos el título del gráfico
plt.show()
#Explicación de las posiciones en un vector
x = np.arange(-15,15+0.5,0.5) #arreglo que contiene el rango de interés
y = 3*x**2+7*x+2 #función trayectoria del satélite en función de x
print(y[0]) #se muestra el primer elemento de la función trayectoria
print(y[15:20]) #se muestran los elementos del 15 al 20 de y
print(y[4::3]) #se muestran los elementos de y empezando en 4 y va en pasos de 3
```

**Punto #4: Definiendo otras trayectorias: Definición de funciones**

Se muestran los gráficos de un par de funciones que represemtan trayectorias de otros satélites de la red.

```python
#Definiendo otras trayectorias de otros satélites
import numpy as np
from matplotlib import pyplot as plt

t = np.arange(-15,30.1,0.1) #arreglo que contiene el rango de interés
x = 15*np.cos(np.pi*t) #función que representa trayectorias de satélites
plt.figure()
plt.plot(t,x,color='purple')
plt.grid()
plt.xlabel('t'), plt.ylabel('x') #definimos los ejes del gráfico
plt.title('Trayectoria de algunos satélites') #definimos el título del gráfico
plt.show()
```

```python
#Definiendo otras trayectorias de otros satélites
import numpy as np
from matplotlib import pyplot as plt

t = np.arange(-0.10,0.251,0.001) #arreglo que contiene el rango de interés
y = np.exp((5*t))*np.sin(2*t) #función que representa trayectorias de satélites
plt.figure()
plt.plot(t,y,color='purple')
plt.grid()
plt.xlabel('t'), plt.ylabel('y') #definimos los ejes del gráfico
plt.title('Trayectoria de algunos satélites') #definimos el título del gráfico
plt.show()
```

 **Punto #5. Satélite geoestacionario: Funciones polinomiales**

Un satélite geoestacionario posee esta posición su trayectoria: -3, 1, 5 + j, 7 - j

A continuación se muestran sus posiciones en el plano complejo.

```python
#Satélite geoestacionario
import numpy as np
from matplotlib import pyplot as plt
z = np.array([-3,1,5+1j,7-1j]) #arreglo que contiene las posiciones del satélite
z.real
z.imag

plt.xlabel('Real'), plt.ylabel('Imaginario') #Definimos los ejes de la gráfica
plt.plot(z.real, z.imag, 'o', color='purple')
plt.title('Posiciones del satélite geoestacionario')
plt.grid() #Ponemos una malla detrás de la gráfica
plt.show
```
