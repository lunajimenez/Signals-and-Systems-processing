**II. Marco teórico.**

Una **señal es analógica** cuando su amplitud tiene la capacidad de tomar un número infinito de valores en un rango de tiempo continuo. Cuando nos referimos a señales análogas, estamos hablando de señales producidas por la propia naturaleza, de manera más específica, un fenómeno electromagnético. A su vez, estas señales son representadas por una función matemática continua. Normalmente, podemos encontrar señales análogas entre algunas magnitudes físicas, que pueden ser representadas en valores como intensidad, tensión o potencia. También pueden ser hidráulicas, como la presión, o térmicas, como la temperatura.

Dicho de otra manera, las señales analógicas son datos que pasan continuamente de un valor a otro en todos los valores intermedios. Por lo tanto, su variación se mantiene a lo largo del tiempo.  
¿Cuál es la forma en que se perciben las señales analógicas?
En realidad, es posible detectar señales analógicas en casi cualquier lugar. ¿Cuál es el propósito de esto? Es muy simple: la naturaleza tiene un conjunto de señales como la luz, la energía, el sonido, etc., de las cuales se pueden tomar todos los valores posibles de un intervalo.

Por otro lado, la **señal digital** siempre tiene valores muy discretos porque se representa mediante una variación discontinua con el tiempo. ¿Qué significa esto? simplemente, porque solo puede tomar dos valores, la información no se transmite de la misma manera. Como utiliza un sistema de código binario, solo acepta 0 y 1.

A pesar de que solo se trata de dos valores, la transmisión digital bajo una pareja de amplitudes ofrecerá grandes posibilidades.
Los pulsos, que son las representaciones de la señal digital básica en el dominio del tiempo, son una onda cuadrada. La duración (ancho de pulso), la frecuencia de repetición (velocidad de pulso por segundo) y la altura de pulso (magnitud eléctrica) son sus parámetros de medición.
La manera en la que se perciben las señales digitales es completamente distinta a las analógicas a pesar de que se trata de una forma de interpretar magnitudes. Resulta evidente que el mundo físico no produce estas señales como tal, sino que el hombre las crea. Estas son señales que el hombre ha tomado durante un método de tratamiento. Como se mencionó anteriormente, la señal es una onda cuadrada y su representación se realiza en el dominio del tiempo. [1]

**Propagación del sonido en el aire**

El sonido es un fenómeno físico que se define como la propagación de una deformación en un medio elástico (sólido, líquido o gaseoso). La propagación del sonido se produce en forma de ondas longitudinales de presión debido a las fluctuaciones volumétricas y de presión locales, lo que lo convierte en un proceso termodinámico. Debido a que las ondas sonoras se consideran un fenómeno termodinámico, es posible examinar qué tipo de procesos están involucrados en la propagación de este fenómeno. Como se mencionó anteriormente, la propagación de las ondas sonoras implica cambios locales de volumen y presión, los cuales están relacionados y explicados por solo dos procesos termodinámicos.  [2]

**Conversión análoga a digital**

Para adquirir señales sonoras se emplea un micrófono. Un micrófono funciona como un transductor y transforma la energía mecánica (producida por el movimiento del diafragma) en una señal eléctrica.  Luego, la tarjeta de sonido de la computadora convierte esta señal en un proceso de conversión analógico-digital (A/D).
El muestreo discretiza el eje temporal y cuantifica la amplitud de las señales durante el proceso de conversión analógico-digital (A/D).
La resolución de la conversión A/D y, por lo tanto, la calidad de la señal producida está determinadas por el número de bits utilizados en la codificación. Un mayor número de bits permite una representación de la señal original más precisa y fiel. El paquete IPython de Python permite grabar y reproducir señales de audio.

```python
# Importamos las librerías NumPy, SciPy y Matplotlib, estas permiten generar
# manipular y graficar señales, y las funciones para el procesamiento de audio.
from matplotlib import pyplot as plt
import numpy as np
from scipy import signal
from IPython.display import Audio
```

**Punto 1.** La función signal.chirp permite generar un tono de frecuencia variable. Se hizo uso de esta función para generar un tono de dos segundos de duración y frecuencias entre 100Hz y 700Hz. También se muestra la gráfica de la señal generada.

El sonido que se obtuvo va aumentando de frecuencia a medida que pasa el tiempo, empieza sin que pueda ser percibido y aumenta.


```python
#En el siguiente ejemplo se muestra cómo generar la señal chirp para frecuencias de 100Hz
#a 700Hz con una duración de 5 segundos:

#@title
fs=4000
duracion=10 #<- segundos
t = np.linspace(0,duracion,duracion*fs) #cantidad de elementos es duración*fs y calcula el paso
w = signal.chirp(t, f0=100, f1=700, t1=10, method='linear') #tiempo 0 luego su frecuencia. Luego frecuencia y luego su tiempo.
plt.figure(figsize=(50,10))
plt.plot(t,w)
plt.title("Linear Chirp, f(0)=100, f(10)=70")
plt.xlabel('t(sec)')
plt.show()
#Comenzar reproducción
Audio(w, rate=fs)
```

¿Se podría generar el sonido de la sirena de una ambulancia con Python? La respuesta es sí y se prueba a continuación.

```python
#Haciendo un sonido parecido al de una ambulancia.
#Para esto variamos la frecuencia f1.
#@title
fs=4000
duracion=10 #<- segundos
t = np.linspace(0,duracion,duracion*fs) #cantidad de elementos es duración*fs y calcula el paso
w = signal.chirp(t, f0=100, f1=9000, t1=10, method='linear') #tiempo 0 luego su frecuencia. Luego frecuencia y luego su tiempo.
plt.figure(figsize=(50,10))
plt.plot(t,w)
plt.title("Linear Chirp, f(0)=100, f(10)=70")
plt.xlabel('t(sec)')
plt.show()
#Comenzar reproducción
Audio(w, rate=fs)
```

**Punto 2.** El ruido en un sistema de sonido puede modelarse empleando señales aleatorias. A continuación se muestra cómo se generó un ruido aleatorio con distribución de probabilidad normal de la misma duración y se reprodujo a la misma frecuencia del ejemplo anterior.

```python
#Ruido es cualquier tipo de interferencia externa a mi sistema. Una perturbación a mi sistema.
#el ruido tiene un comportamiento más o menos predecible

#crea un conjunto de datos que simulan la campana de Gauss
noise = np.random.randn(duracion*fs)
plt.figure(figsize=(50,10)), plt.plot(t,noise), plt.show()
Audio(noise,rate=fs)
```

**Punto #3.** A continuación se encuentra el audio de la señal resultante de la suma de la señal chirp y el ruido. Este resultado es la combinación de ambas señales y se alcanzan a apreciar las dos. Al principio solo es posible percibir el ruido, pero a medida que avanza el tiempo se empieza a escuchar la señal chirp y esta va aumentando su frecuencia hasta el final del audio.

```python
#Sonido
fs=4000
duracion=10 #<- segundos
t = np.linspace(0,duracion,duracion*fs) #cantidad de elementos es duración*fs y calcula el paso
w = signal.chirp(t, f0=100, f1=700, t1=10, method='linear') #tiempo 0 luego su frecuencia. Luego frecuencia y luego su tiempo.

#Ruido
noise = np.random.randn(duracion*fs)

#Sonido + ruido
y = w+noise

#Graficando
plt.figure(figsize=(50,10)), plt.plot(t,y), plt.show()
Audio(y,rate=fs)
```

**Punto #4.** Se hizo uso de la función fijada en SAVIO para grabar mi voz pronunciando una palabra o frase corta durante 4 segundos o menos.

```python
#Función con dos salidas. Se tiene una nueva frecuencia de muestreo llamada sr.
#@title Función grabar

from IPython.display import HTML, Audio
from google.colab.output import eval_js
from base64 import b64decode
import numpy as np
from scipy.io.wavfile import read as wav_read
import io
import ffmpeg
AUDIO_HTML = """
<script>
var my_div = document.createElement("DIV");
var my_p = document.createElement("P");
var my_btn = document.createElement("BUTTON");
var t = document.createTextNode("Press to start recording");
my_btn.appendChild(t);
//my_p.appendChild(my_btn);
my_div.appendChild(my_btn);
document.body.appendChild(my_div);
var base64data = 0;
var reader;
var recorder, gumStream;
var recordButton = my_btn;
var handleSuccess = function(stream) {
  gumStream = stream;
  var options = {
    //bitsPerSecond: 8000, //chrome seems to ignore, always 48k
    mimeType : 'audio/webm;codecs=opus'
    //mimeType : 'audio/webm;codecs=pcm'
  };
  //recorder = new MediaRecorder(stream, options);
  recorder = new MediaRecorder(stream);
  recorder.ondataavailable = function(e) {
    var url = URL.createObjectURL(e.data);
    var preview = document.createElement('audio');
    preview.controls = true;
    preview.src = url;
    document.body.appendChild(preview);
    reader = new FileReader();
    reader.readAsDataURL(e.data);
    reader.onloadend = function() {
      base64data = reader.result;
      //console.log("Inside FileReader:" + base64data);
    }
  };
  recorder.start();
  };
recordButton.innerText = "Recording... press to stop";
navigator.mediaDevices.getUserMedia({audio: true}).then(handleSuccess);

function toggleRecording() {
  if (recorder && recorder.state == "recording") {
      recorder.stop();
      gumStream.getAudioTracks()[0].stop();
      recordButton.innerText = "Saving the recording... pls wait!"
  }
}
// https://stackoverflow.com/a/951057
function sleep(ms) {
  return new Promise(resolve => setTimeout(resolve, ms));
}
var data = new Promise(resolve=>{
//recordButton.addEventListener("click", toggleRecording);
recordButton.onclick = ()=>{
toggleRecording()
sleep(2000).then(() => {
  // wait 2000ms for the data to be available...
  // ideally this should use something like await...
  //console.log("Inside data:" + base64data)
  resolve(base64data.toString())
});
}
});

</script>
"""
def get_audio():
  display(HTML(AUDIO_HTML))
  data = eval_js("data")
  binary = b64decode(data.split(',')[1])

  process = (ffmpeg
    .input('pipe:0')
    .output('pipe:1', format='wav')
    .run_async(pipe_stdin=True, pipe_stdout=True, pipe_stderr=True, quiet=True, overwrite_output=True)
  )
  output, err = process.communicate(input=binary)

  riff_chunk_size = len(output) - 8
  # Break up the chunk size into four bytes, held in b.
  q = riff_chunk_size
  b = []
  for i in range(4):
      q, r = divmod(q, 256)
      b.append(r)
  # Replace bytes 4:8 in proc.stdout with the actual size of the RIFF chunk.
  riff = output[:4] + bytes(b) + output[8:]
  sr, audio = wav_read(io.BytesIO(riff))
  return audio, sr
```

Posteriormente, se emplea la siguiente línea de
código para llamar la función y empieza la grabación de mi voz.


```python
audio, sr = get_audio() #grabación del audio
```

**Punto #5.**

Comprimiendo en el tiempo por un factor de 4.

```python
#Para hacerlo se debe generar un nuevo vector que contenga solo la cuarta parte de elementos del vector original.
#Se utilizó la técnica de slicing audio[::4], se está seleccionando solo una cuarta parte de los elementos del vector original.
#Esto equivale a comprimir la señal en el tiempo por un factor de 4,
#ya que se están tomando muestras a una frecuencia de muestreo reducida en un factor de 4.
audio_comprimido = audio[::4]
# Reproducir el nuevo audio comprimido en el tiempo por un factor de 4
Audio(audio_comprimido, rate=sr)
```

Reflejar en el tiempo. La señal se reproducirá a continuación en el orden inverso.


```python
#Para este caso se utilizó la función np.flipud, esta función
#es de la biblioteca NumPy para invertir el orden de los elementos en la señal de audio.
audio_reflejado = np.flipud(audio)
# Reproducir el nuevo audio reflejado en el tiempo.
Audio(audio_reflejado, rate=sr)
```

Ejecutar la señal a la mitad de la frecuencia de muestreo.

```python
sr_mitad = (sr/2) #se reduce la frecuencia de muestreo a la mitad
audio_mitad_frecuencia = audio[::2] #se toma cada segundo elemento de la señal de audio, se reducen la cantidad de muestras a la mitad y esto reduce su frecuencia.
# Reproducir el nuevo audio a la mitad de la frecuencia de muestreo
Audio(audio_mitad_frecuencia, rate=sr)
```

**Parte 3. Generación de eco.**

El eco es una señal desplazada en el tiempo con reducción de su amplitud.

**Punto #6.**

Se generará un eco de la señal de voz original. Para ello, debe
crearse una señal que sea una versión a escala de la original y sumarla a ésta con un retardo de tiempo.

El siguiente código muestra cómo definir una versión escalada y retrasada de la señal. Tenga en cuenta que los ceros se insertan al comienzo de la señal.

```python
#Primero debemos reducir la amplitud de la señal
#Luego, multiplicamos la señal por valor inferior a 1. Para que sea menor el sonido que al de la señal original.
#Para desplazar la señal, mi variable t la desplazo restándole un valor T para que se escuche después

a = 0.5 # attenuation
delay = round(0.2*sr) # a 0.2-second delay
xe = np.append(np.zeros(delay), a*audio[:-delay])
rt = audio + xe
Audio(rt,rate=sr)
```

A continuación se variaron los valores de la atenuación y del delay (retardo).

Las diferencias en el sonido obtenido a continuación radican en que al aumentar la atenuación se reduce la amplitud de la señal de eco en relación con la señal original, lo que hace al eco escucharse de manera más suave. Por otro lado, al aumentar mucho el delay, el eco se sobrepondrá sobre el audio original.

```python
a = 0.8 # attenuation
delay = round(0.9*sr) # a 0.9-second delay
xe = np.append(np.zeros(delay), a*audio[:-delay])
rt = audio + xe
Audio(rt,rate=sr)
```
